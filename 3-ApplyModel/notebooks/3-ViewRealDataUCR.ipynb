# Anomaly benchmark UCR archive

```python
#import libraries
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
import re 
import os 
import seaborn as sns
sns.set_theme()
```

```python
BASE_DIR=".."
data_dir=os.path.join(BASE_DIR,"data2apply/UCR_dataAnomalyArchive")
list_files=os.listdir(data_dir)
list_data_files=[ os.path.join(data_dir,i) for i in list_files]



indexes=[]
for j in list_files:   
    indexes.append([int(i.split(".")[0]) for i in j.split("_")[-3:]])



dir_data_files={int(i.split("_")[0]):{"path":j,"index":ind} for i,j,ind in zip(list_files,list_data_files,indexes)}
```

```python
plt.figure(figsize=(19,8))

for ID in range(1,7):
    plt.subplot(2,3,ID)
    aux=np.loadtxt(dir_data_files[ID]["path"])
    train,start,end=dir_data_files[ID]["index"]

    plt.plot(aux[(start-400):(end+400)])
    plt.axvspan(400,400+(end-start),alpha=0.2,color="red")
    plt.title(f"ID: {ID}",fontsize=14)
    plt.tight_layout()
    plt.savefig("../../visualizaciones_tfm/imgs/UCR_exmaple.png")
```
